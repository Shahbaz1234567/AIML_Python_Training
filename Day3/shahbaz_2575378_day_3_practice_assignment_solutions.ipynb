```python
''' PS3: write the function called sed that takes as arguments a pattern string, a replacement string,and two filenames; it should
read the first file and write the contents into second file(creating if necessary ). If the pattern string appears anywhere in 
the file it should be replaced with the replacement string. If an error occurs while opening,reading,writing,or closing files,
your program should be catch the exception, print an error message,and exit '''

import os
import sys

def sed(pattern, replacement, first, second):
    try:
        fi = open(first)
        fo = open(second, 'w+')

        for line in fi:
            if pattern in line:
                print('pattern found')
                
                newLine = line.replace(pattern,replacement)
                fo.write(newLine)
            else:
                fo.write(line)
    except:
        print(sys.exc_info()[0].__name__)
    finally:
        f.close()
        print("Exit")


pattern = 'op'
replacement = 'MPHASIS'
first = "F:/AIML_Mphasis/AIML_Python/Day3/f1.txt"
second = "F:/AIML_Mphasis/AIML_Python/Day3/f2.txt"
sed(pattern, replacement, first, second)

```

```python
''' PS#4: Log File Analysis: You have a log file containing records of user activities on a website. Each line in the file 
represents a log entry with details like timestamp,userID,and action performed. Your task is to analyze this log file. 

'''
import re

#log_file_path = "C:/Users/Public/Documents/AdobeGCData/adobegc.log"
log_file_path = "C:/Users/shabaz/Downloads/logfile.txt"

def parse_log_entry(entry):
    parts = re.split(r'\s+', entry)
    if len(parts) >= 3:
        timestamp, user_id, action = parts[:3]
        return timestamp, user_id, action
    else:
        return None

def analyze_log(log_file_path):
    actions_count = {}  
    peak_hour = None
    peak_hour_count = 0

    try:
        with open(log_file_path, 'r') as log_file:
            for line in log_file:
                timestamp, user_id, action = parse_log_entry(line)
                if timestamp and action:
                    if action in actions_count:
                        actions_count[action] += 1
                    else:
                        actions_count[action] = 1

                    
                    hour = timestamp.split(':')[0]
                    if actions_count[action] > peak_hour_count:
                        peak_hour = hour
                        peak_hour_count = actions_count[action]

        
        print("Action Counts:")
        for action, count in actions_count.items():
            print(f"{action}: {count} times")

        print(f"Peak Hour for Action '{peak_hour}': {peak_hour_count} times")
    
    except FileNotFoundError:
        print(f"Log file not found: {log_file_path}")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    analyze_log(log_file_path)

```

```python
'''PS#5: Text File Search and Replace: You have text file with large amount of text, and you want to search for specific  words 
or phrases and replace them with new content.
  a. write the python code to search and replace text within a text file.
  b. How would you handle cases where you need to perform multiple replacements in a single pass?

'''

import re

def search_replace(file_path, list1):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
        
        for search_pattern, replace_text in list1:
            content = re.sub(search_pattern, replace_text, content)
        
        with open(file_path, 'w') as file:
            file.write(content)
        
        print(f"Search and replace completed in {file_path}")
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

def result():
    file_path = input("Enter the path to the text file: ")
    list1 = []

    while True:
        search_pattern = input("Enter the text or regular expression to search for (leave empty to finish): ")
        if not search_pattern:
            break
        replace_text = input("Enter the replacement text: ")
        list1.append((search_pattern, replace_text))

    if not list1:
        print("No search and replace patterns provided. Exiting.")
        return

    search_replace(file_path, list1)

if __name__ == "__main__":
    result()
```

```python
#PS6. write a python script that concatenates the contents of multiple text files into a single outpul file. Allow the user to 
#    specify the input files and the output file.

def concatenate_files(input_files, output_file):
    try:
        with open(output_file, 'w') as output:
            for input_file in input_files:
                with open(input_file, 'r') as f:
                    output.write(f.read())
        print(f"Concatenated {len(input_files)} files into {output_file}")
    except FileNotFoundError:
        print("One or more input files not found.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

def result():
    input_files = []
    while True:
        file_name = input("Enter an input file (leave empty to finish): ")
        if not file_name:
            break
        input_files.append(file_name)

    if not input_files:
        print("No input files provided. Exiting.")
        return

    output_file = input("Enter the output file: ")

    concatenate_files(input_files, output_file)

if __name__ == "__main__":
    result()
```

```python
''' PS7:  you are given a text file named input.txt containing a list of words, one word per line. your task is to create a python
program that reads the contents of input.txt,processes the words,and write the result to an output file named output.txt.
     The program should perform the following operations:
     i. Read the words from input.txt
     ii. For each word in input file, calculate the length of the word and store it in a dictionary where the word is the key,
         and the length is the value.
     iii. write the word-length dictionary to output.txt in the following format:   '''



def word_length_dict(input_filename):
    
    dict_len = {}
    try:
        with open(input_filename, 'r') as input_file:
            for line in input_file:
                word = line.strip()
                length = len(word)
                dict_len[word] = length
        return dict_len
    except FileNotFoundError:
        print(f"Error: The input file '{input_filename}' was not found.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None


def dict_to_file(dict_len, output_filename):
    try:
        with open(output_filename, 'w') as output_file:
            for word, length in dict_len.items():
                output_file.write(f"{word}: {length}\n")
        print(f"Word-length dictionary has been written to '{output_filename}'.")
    except PermissionError:
        print("Error: Permission denied to write to the output file.")
    


input_filename = "input.txt"
output_filename = "output.txt"

dict_len = word_length_dict(input_filename)
if dict_len is not None:
    dict_to_file(dict_len, output_filename)

```

```python
''' PS8: Assume that you are developing a student gradebook system for a school. The system should allow teachers to input 
student grades for various subjects, store the data into files,and provide students with the ability to view their grades.
    Design a python program that accomplishes the following tasks:
      i. Teachers should be able to input grades for students in different subjects.
      ii. Store the student grade data in separate text files for each subjects.
      iii. Students should be able to view their grades for each subjects.
      iv. Impelement error handling for file operations, such as file not found or permission issues. '''

import os

def input_grades(subject):
    try:
        filename = f"{subject}.txt"
        with open(filename, 'a') as file:
            while True:
                student_name = input(f"Enter student name for {subject} (or 'q' to quit): ")
                if student_name == 'q':
                    break
                grade = input(f"Enter {student_name}'s grade for {subject}: ")
                file.write(f"{student_name}: {grade}\n")
        print(f"Grades for {subject} have been successfully recorded.")
    except IOError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


        
def view_grades(subject):
    try:
        filename = f"{subject}.txt"
        with open(filename, 'r') as file:
            grades = file.read()
            print(f"Grades for {subject}:\n{grades}")
    except FileNotFoundError:
        print(f"Grades for {subject} have not been recorded yet.")
    except PermissionError:
        print("Permission denied to access the file.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


while True:
    print("\nOptions:")
    print("1. Input grades for a subject")
    print("2. View grades for a subject")
    print("3. Quit")

    choice = input("Enter your choice (1/2/3): ")

    if choice == '1':
        subject = input("Enter the subject: ")
        input_grades(subject)
    elif choice == '2':
        subject = input("Enter the subject: ")
        view_grades(subject)
    elif choice == '3':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please select 1, 2, or 3.")

```
