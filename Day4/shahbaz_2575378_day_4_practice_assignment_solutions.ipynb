```python
'''PS#1: write a pyhton program with exception handling to input marks for five 
subjects Physics, Chemistry, Biology, Mathematics and Computer. calculate the 
percentage and grade according to the following:
 i) Percentage>=90%: Grade A
 ii)Percentage>=90%: Grade B
 iii)Percentage>=90%: Grade C
 iv) Percentage>=90%: Grade D
 v) Percentage>=90%: Grade E
 vi) Percentage>=90%: Grade F
 
'''

def input_marks(marks):
    total_marks = sum(marks)
    percentage = (total_marks / 500) * 100

    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    elif percentage >= 40:
        return 'E'
    else:
        return 'F'

try:
    subjects = ["Physics", "Chemistry", "Biology", "Mathematics", "Computer"]
    marks = []

    for subject in subjects:
        mark = float(input(f"Enter marks for {subject}: "))
        if mark < 0:
            raise ValueError("Your marks is less than 0; Invalid marks")
        elif mark > 100:
            raise ValueError("Your marks is greater than 100; Invalid marks")    
        marks.append(mark)

    grade = input_marks(marks)
    print(f"Percentage: {sum(marks)/5:.2f}%")
    print(f"Grade: {grade}")
except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
'''PS#2: write a python program with exception handling to input electricity unit charges and calculate the total electricity bill
according to the given condition:
  i) For the first 50 units Rs 0.50/unit
  i) For the next 100 units Rs 0.75/unit
  i) For the next 100 units Rs 1.20/unit
  i) For the units above 250 Rs 1.50/unit
  v) An additional surcharge of 20% is added to the bill.

'''
import sys

try:
    units = float(input("Enter the electricity units consumed: "))
    if units < 0:
        raise ValueError("Units cannot be negative")

    if units <= 50:
        total_bill = units * 0.50
    elif units <= 150:
        total_bill = (50 * 0.50) + ((units - 50) * 0.75)
    elif units <= 250:
        total_bill = (50 * 0.50) + (100 * 0.75) + ((units - 150) * 1.20)
    else:
        total_bill = (50 * 0.50) + (100 * 0.75) + (100 * 1.20) + ((units - 250) * 1.50)

    surcharge = 0.20 * total_bill
    total_bill += surcharge

    print(f"Total electricity bill: Rs. {total_bill:.2f}")
except ValueError as e:
    print(f"Error: {e}")
    print(sys.exc_info()[0].__name__)
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    print(sys.exc_info()[1])

print('End')
```

```python
# PS3: write a python program to Exception handling to input the week number and print the weekday.
import sys

try:
    wn = int(input("Enter the DayNumber in the week: "))
    if wn < 1:
        raise ValueError("day number is not less than 1")
    elif wn > 7:
        raise ValueError("day number is not greater than 7")    

    weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    weekday = weekdays[wn - 1]

    print(f"Day of the week for week number {wn}: {weekday}")
except ValueError as e:
    print(f"Error: {e}")
    print(sys.exc_info()[2])
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    print(sys.exc_info()[0].__name__)
    
    
print("End:")    

```

```python
'''PS#5: write the python program to find most frequent words in a text read from file.
        i) Initially open the text file in read mode.
        ii) Make all the letters in the document into lowercase letters and split the words in each line.
        iii) Get the words in an order.
        iv) Sort the words for finding the most frequent words in the file.
        v) Print the most frequent words in the file.

'''

import sys
from collections import Counter

fname = "myfile.txt"  
try:
    with open(fname, "r") as file:
        text = file.read().lower()
        words = text.split()

        word_count = Counter(words)

        # Sort words by frequency (most frequent first)
        sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

        # Print the most frequent words
        num_most_frequent = 7  # Change this value to print a different number of words
        print(f"Top {num_most_frequent} most frequent words:")
        for word, frequency in sorted_words[:num_most_frequent]:
            print(f"{word}: {frequency}")
except FileNotFoundError:
    print(f"The file '{fname}' was not found.")
    print(sys.exc_info()[0].__name__)
except Exception as e:
    print(f"An error occurred: {e}")
    print(sys.exc_info()[1])
    
    
print("\n\nEnd:")    
```
